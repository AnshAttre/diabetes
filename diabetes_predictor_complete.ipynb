```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.svm import SVC
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_data=pd.read_csv(r"C:\Users\Ansh\Downloads\diabetes.csv")
diabetes_data.tail()
```

```python
diabetes_data.shape
```

```python
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()
```

```python
diabetes_data.groupby('Outcome').mean()
```

```python
x=diabetes_data.drop(columns='Outcome',axis=1)
y=diabetes_data['Outcome']
```

```python
x
```

```python
y
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data=scaler.transform(x)
```

```python
standardized_data
```

```python
x=standardized_data
y=diabetes_data['Outcome']
```

```python
x
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,X_train.shape,X_test.shape)
```

# Training model

# SVM

```python
svm_classifier= svm.SVC(C=0.1,kernel='poly',degree=2,gamma=0.1,coef0=0.5)
```

```python
svm_classifier.fit(X_train,Y_train)
```

```python
X_train_prediction=svm_classifier.predict(X_train)
```

```python
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
training_data_accuracy
```

```python
svm_classifier.fit(X_test,Y_test)
```

```python
testing=svm_classifier.predict(X_test)
```

```python
testing_accuracy=accuracy_score(testing,Y_test)
```

```python
testing_accuracy
```
